```python
import numpy as np
import pandas as pd
```

```python
X=pd.read_csv('Iris-new.csv')
X.head()
```

```python
y=X.Species
y = np.array(y)
x=X.drop('Species',axis=1)
x=x.drop('Id',axis=1)
x=np.array(x)
print(y)
print(x)
```

```python
seedNo = 11
np.random.seed(seedNo) # for reproducibility

"""------------------------------------------------------------------
Type III: Kfold training with k= 5 
---------------------------------------------------------------------"""
 
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import StratifiedKFold # K-fold cross validation - Accuarate but huge computation


# define 5-fold cross validation
kfold = StratifiedKFold(n_splits = 5, shuffle=True, random_state = seedNo)

score_acc = []
score_loss = []
for train, test in kfold.split(x, y):
    # Define model
    mykeras_model = Sequential()
    mykeras_model.add(Dense(10, input_dim = 4, activation = 'relu'))
    mykeras_model.add(Dense(12, activation = 'relu'))
    mykeras_model.add(Dense(1, activation = 'sigmoid'))
    
    
    # Compile model
    mykeras_model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
    
    
    # Fit the model - 100 epoch - manual train test
    mykeras_model.fit(x[train], y[train], epochs = 20, batch_size = 8, verbose = 0)
    
    
    # Evaluate model
    perf_meas = mykeras_model.evaluate(x[test], y[test], verbose = 2)    
       
    
#    print("\n")
    print("100 Epoch - 10 fold Results: %s: %f%%" % (mykeras_model.metrics_names[1], perf_meas[1]*100))

##################################  
```
